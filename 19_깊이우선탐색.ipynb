# **1. 깊이 우선 탐색(Depth-First Search)**

### 1-1. DFS란?
- 대표적인 그래프 탐색 알고리즘
- 정점의 자식들을 먼저 탐색하는 방식
- 한 노드의 자식을 타고 끝까지 순회한 후, 다시 돌아와서 다른 형제들의 자식을 타고 내려가며 순회함

```python
graph =  dict()

# 키 값이 갈 수 있는 노드들의 정보를
# 딕셔너리 형태로 저장했습니다.
# 간선이 이렇게 표시되어있다는 것을 유추할 수 있다.
graph['A'] = ['B', 'C']
graph['B'] = ['A', 'D']
graph['C'] = ['A', 'G', 'H', 'I']
graph['D'] = ['B', 'E', 'F']
graph['E'] = ['D']
graph['F'] = ['D']
graph['G'] = ['C', 'H']
graph['H'] = ['G', 'C', 'I']
graph['I'] = ['H', 'C', 'J']
graph['J'] = ['I']
```

# **2. DFS 알고리즘 구현**
- 스택과 큐 자료구조를 활용
- need_visit 스택, vistied 큐 생성

- 깊이 우선 탐색 결과
    - A, C, I, J, H, G, B, D, F, E

```python
def dfs(graph, start_node) :
    # visited : 방문했던 장소
  # need_visit : 방문이 필요한 장소
  visited, need_visit = list(), list()

  need_visit.append(start_node)

  while need_visit:
    node = need_visit.pop()
    if node not in visited:
      visited.append(node)
      need_visit.extend(graph[node])

  return visited
```

```python
dfs(graph, 'A')
```
