# **1. 더블링크드리스트(Doubly Linked List)**
- 양방향으로 연결되어 있어서 노드 탐색이 양쪽으로 모두 가능

````python
class Node:
  def __init__(self, data, prev = None, next = None): # 이전, 다음 주소
    self.prev = prev
    self.data = data
    self.next = next
````

````python
class NodeMgmt:
  # 더블링크드리스트를 만들 때, 노드 하나만 넣어주고,
  # 그 노드가 헤드이자 테일이 되도록 한다.
  def __init__(self, data):
    self.head = Node(data)
    self.tail = self.head

  def insert(self, data):
    # 방어코드
    if self.head == None:
      self.head = Node(data)
      self.tail = self.head
    else:
      node = self.head
      while node.next:
        node = node.next
      # 변수에 담아놓지 않고 계속 생성자를 호출하면
      # 우리가 원하는대로 작동하지 않는다.
      new = Node(data)
      node.next = new
      new.prev = node
      self.tail = new

  def print_node(self):
    node = self.head
    while node:
      print(node.data)
      node = node.next

  # 문제1
  def insert_before(self, data, after_data): # 2.5 3
    if self.head == None:
      self.head = Node(data)
      self.tail = self.head
      return
    else:
      node = self.tail
      while node.data != after_data:
        if node == None:
          return
        node = node.prev
      new = Node(data)
      before_new = node.prev # 원래 node.prev는 2를 가리키고 있었음, before_new : 2의 주소
      before_new.next = new # 2.5의 주소를 before_new.next에 저장, 2 주소의 next는 2.5를 가리키게 된다.
      new.prev = before_new # 2의 주소를 2.5의 prev에 저장한다. 2.5의 prev는 2를 가리키게 된다.
      new.next = node # 현재 node의 주소를 new.next에 저장한다. 2.5의 next는 3을 가리키게 된다.
      node.prev = new # 2.5의 주소를 3의 prev에 저장한다. 3의 prev는 2.5를 가리키게 된다.

  # 문제2
  def search_from_head(self, data):
    # 이 함수의 기능은 검색이다.
    # 그렇기에 내가 찾는 데이터가 없다하더라도 추가가 아닌
    # 없다라고 False 리턴할 것이다.
    if self.head == None:
      return False
    node = self.head
    while node:
      if node.data == data:
        return node.data
      else:
        node = node.next
    return False

  # 문제3
  def search_from_tail(self, data):
    if self.tail == None:
      return False
    node = self.tail
    while node:
      if node.data == data:
        return node.data
      else:
        node = node.prev
    return False
````

````python
double_linked_list = NodeMgmt(0)

for data in range(1, 11):
  double_linked_list.insert(data)

double_linked_list.print_node()
````
