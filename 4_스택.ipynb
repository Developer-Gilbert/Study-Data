# **1. 스택(stack)**
- 데이터를 제한적으로 접근할 수 있는 구조
- 한쪽 끝에서만 자료를 넣거나 뺄 수 있는 구조
- 가장 나중에 쌓은 데이터를 가장 먼저 빼낼 수 있는 구조
- LIFO(Last Input First Out)

### 1-1. 스택의 구조
- 스택은 LIFO 또는 FILO 데이터 관리 방식
- 스택의 활용 : 컴퓨터 내부의 프로세스 구조의 함수 동작 방식
- 기능
  - push() : 데이터를 스택에 쌓기
  - pop() : 데이터를 스택에서 꺼내기
- [Visualgo](https://visualgo.net/en/list)

### 1-2. 스택의 장단점
- 구조가 단순해서 이해와 구현이 쉬움
- 데이터 저장/읽기 속도가 빠름
- 데이터 최대 개수를 미리 정해야 함(파이썬의 경우 재귀함수는 1000번까지만 호출이 가능)
- 저장 공간의 낭비가 발생할 수 있음(미리 최대 갯수만큼 저장공간을 미리 확보하기 때문)
> 스택은 단순하고 빠른 성능을 위해 사용하므로 보통 배열 구조를 활용해서 구현하는 것이 일반적

```python
# 파이썬 리스트에서 제공하는 메소드로 스택 사용하기
data_stack = list()

data_stack.append(2)
data_stack.append(10)
data_stack.append(49)
print(data_stack)

print(data_stack.pop())
print(data_stack)

# 자료구조에서 데이터를 삭제한다라는 의미는,
# 저장되어있는 데이터를 가지고 나와서 사용한다라는 것을 의미한다.
# 그렇기에 원활하게 사용하기 위해서는 변수에 담아주는게 일반적이다.
temp = data_stack.pop()
print(temp)
print(data_stack)
```

(단, 리스트의 pop 함수는 사용하지 않음)

다 푸신 분은, append도 사용하지 않고 해보기!

```python
stack_list = list()

def push(data):
  # stack_list.append(data)

  # append 사용하지 않고 하기!

  # 첫번째 방법
  global stack_list
  stack_list = stack_list + [data]

  # 두번재 방법
  # stack_list[len(stack_list):] = [data]

def pop():
  data = stack_list[-1]
  del stack_list[-1]
  return data
```

```python
for index in range(10):
  push(index)
```
