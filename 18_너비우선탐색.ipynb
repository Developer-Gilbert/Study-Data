# **1. 너비 우선 탐색(Breadth-First Search)**

### 1-1. BFS란?
- 대표적인 그래프 탐색 알고리즘
- 정점들과 같은 레벨에 있는 노드(형제 노드)들을 먼저 탐색하는 방식
- 한 단계씩 내려가면서 해당 노드와 같은 레벨에 있는 노드들을 먼저 순회함

```python
graph =  dict()

# 키 값이 갈 수 있는 노드들의 정보를
# 딕셔너리 형태로 저장했습니다.
# 간선이 이렇게 표시되어있다는 것을 유추할 수 있다.
graph['A'] = ['B', 'C']
graph['B'] = ['A', 'D']
graph['C'] = ['A', 'G', 'H', 'I']
graph['D'] = ['B', 'E', 'F']
graph['E'] = ['D']
graph['F'] = ['D']
graph['G'] = ['C', 'H']
graph['H'] = ['G', 'C', 'I']
graph['I'] = ['H', 'C', 'J']
graph['J'] = ['I']
```

# **2. BFS 알고리즘 구현**

- 큐 자료구조를 활용
- need_visit 큐, vistied 큐 생성

```python
# 원래 list 객체의 pop이라는 메소드는
# 가장 마지막에 있는 인덱스를 삭제 후 리턴해준다.
data = [1, 2, 3, 4]
print('삭제된 데이터 : ', data.pop())
print(data)
```

```python
# pop(0), pop이라는 메소드에 0 이라는 인덱스 번호를 매개변수로 전달하면
# FIFO 큐로 구현한 셈 사용할 수 있다.
data = [1, 2, 3, 4]
print('삭제된 데이터 : ', data.pop(0))
print(data)
```

```python
# 편하게 + 를 통해서 리스트 연결도 가능하지만,
# 함수를 한 번 사용해볼거에요!
data = [1, 2, 3, 4]
data += [7, 8, 9]
print(data)
```

```python
# extend() 함수를 이용해서, 두 리스트를 붙인다.
data = [1, 2, 3, 4]
data.extend([4, 5])
print(data)
```

- 너비 우선 탐색 결과
    - A, B, C, D, G, H, I, E, F, J

```python
def bfs(graph, start_node):
  # visited : 방문했던 장소
  # need_visit : 방문이 필요한 장소
  visited, need_visit = list(), list()

  need_visit.append(start_node)

  while need_visit:
    node = need_visit.pop(0)
    # need_visit 에서 하나 꺼내와서 visited에 있는지 없는 지 검사
    # 내가 검사하는 노드가 방문 했는지 안했는지 검사
    # 방문을 하지 않았다면 if문 내의 코드 실행
    if node not in visited:
      # 추가 해준 다음
      visited.append(node)
      # need_visit 에다가는 추가한 노드가 갈 수 있는 경로를 extend
      need_visit.extend(graph[node])

  return visited
```

```python
bfs(graph, 'A')
```
